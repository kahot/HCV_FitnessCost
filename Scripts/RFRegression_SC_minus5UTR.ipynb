## Read the feature file

```python
#read current csv as df called features
features = pd.read_csv('../Data/HCV_allfeattures.csv')
```

```python
#remove 5'UTR
features=features[features["5' UTR"]==0]
```

### Edit Features Dataframe and Setting Labels

```python
#fitness costs
labels=np.array(features['Costs'])
```

```python
labels_log=np.log(labels)
```

```python
# Remove the some features from the features data frame
features = features.drop('Costs', axis = 1) # axis 1 refers to the columns
features = features.drop('Cost Label', axis=1)
features = features.drop('Avg_Mutation_Freq', axis = 1)
features = features.drop('pos', axis=1)
features = features.drop('Mutation_Rate', axis=1)
```

### Check correlation of variables

```python
import seaborn as sns
```

```python
fig, ax = plt.subplots(figsize=(20, 20))
sns.heatmap(features.corr(method='pearson'), annot=False,  vmin=0, vmax=1,cmap="YlGnBu");
```

```python
#looks like 'postive AA' is highly correlated with some features: zoom in
fig, ax = plt.subplots(figsize=(15, 15))
sns.heatmap(features.iloc[:,15:25].corr(method='pearson'),annot=True, fmt='0.2f',vmin=0, vmax=1,cmap="YlGnBu");
```

```python
#Positive AA == Basic AA
#remove one of them
features = features.drop('Basic AA', axis = 1) # axis 1 refers to the columns
```

**Split Data into Training and Testing Sets**

```python
#1. startign with log cost
X_train, X_test, y_train, y_test = train_test_split(features, labels_log, test_size = 0.3, random_state = 0)
#train_features, test_features, train_labels, test_labels = train_test_split(features, le.transform(labels_og), test_size = 0.25, random_state = 42)
```

```python
#Total number of features tested
X_train.shape[1]
```

## Run the regression as default

```python
# Instantiate model with 1000 decision trees
rfr = RandomForestRegressor(n_estimators = 100, random_state = 0)
# Train the model on training data
rfr.fit(X_train, y_train);
print('R^2 Training Score: {:.2f} \nR^2 Testing Score: {:.2f}'.format(rfr.score(X_train, y_train), rfr.score(X_test, y_test)))
```

```python
from pprint import pprint
pprint(rfr.get_params())
```

### Hyperparameter tuning with GridSearchCV

```python
### Pick the parameters to test with GridSearchCV
param_grid = {
    'bootstrap': [True],
    'min_samples_split': [75], 
    'n_estimators' : [200,250],
    'max_depth': [40,50],
    'max_features': [45],
    'min_samples_leaf': [2],    
    }
```

```python
rf = RandomForestRegressor()

# Instantiate the grid search model
grid_search = GridSearchCV(estimator = rf, param_grid = param_grid, 
                          cv = 3, n_jobs = -1, verbose = 2)
```

```python
# Fit the random search model
grid_search.fit(X_train, y_train)
```

```python
print(grid_search.best_params_)
```

```python
print(grid_search.best_params_)
```

```python
def evaluate(model, test_features, test_labels):
    predictions = model.predict(test_features)
    errors = abs(predictions - test_labels)
    mape = 100 * np.mean(errors / test_labels)
    accuracy = 100 - mape
    print('Model Performance')
    print('Average Error: {:0.4f} degrees.'.format(np.mean(errors)))
    print('Accuracy = {:0.2f}%.'.format(accuracy))
    
    return accuracy
```

```python
base_model = RandomForestRegressor(n_estimators = 100, random_state = 42)
base_model.fit(X_train, y_train)
base_accuracy = evaluate(base_model, X_test, y_test)
```

```python
best_grid = grid_search.best_estimator_
grid_accuracy = evaluate(best_grid, X_test, y_test)
```

```python
print('R^2 Training Score: {:.4f} \nR^2 Testing Score: {:.4f}'.format(grid_search.score(X_train, y_train), grid_search.score(X_test, y_test)))
```

### RandomizedSearchCV

### Run the best model

```python
Model1 = RandomForestRegressor(bootstrap=True, 
                            max_depth=40,max_features=45,
                            min_samples_leaf=2,min_samples_split=75,
                            n_estimators=250, n_jobs=-1)
#Fit model
Model1.fit(X_train, y_train)
```

```python
y_pred = Model1.predict(X_test.values)
```
